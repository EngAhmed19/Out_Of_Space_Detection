```python
model2=YOLO("Out_Of_stock.pt")
```

```python
import cv2
```

```python
video_path="shlves2.avi"
output_video_path="Output3.avi"
cap=cv2.VideoCapture(video_path)

frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))

# Define the codec and create a VideoWriter object
fourcc = cv2.VideoWriter_fourcc(*'XVID')  # Codec for .avi files  # Codec for .mp4 files
out = cv2.VideoWriter(output_video_path, fourcc, fps, (frame_width, frame_height))

while True:
    ret,frame=cap.read()

    if not ret:
        break

    results=model2.predict(source=frame,conf=0.5)
    annotaded_frame=results[0].plot()

    cv2.imshow("YOLOv8 detection for out_of_stock",annotaded_frame)

    out.write(annotaded_frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
